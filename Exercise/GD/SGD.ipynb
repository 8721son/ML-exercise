```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(100)
y = 0.3*x + 0.8
print(x)
```

```python
def plot_prediction(pred,y):
    plt.plot(x,y)
    plt.plot(x,pred)
    plt.show()
```

```python
w = np.random.uniform(-1.0,1.0)
bias = np.random.uniform(-1.0,1.0)

num_epoch=1000
learning_rate=0.5
cnt =0
for epoch in range(num_epoch):
    
    x_sgd = np.random.choice(x)
    y_sgd = 0.3*x_sgd + 0.8
    
    predict_sgd = w * x_sgd + bias
    
    w = w - 2*learning_rate*((predict_sgd - y_sgd)*x_sgd)
    bias = bias - 2*learning_rate * (predict_sgd - y_sgd)
    
    predict = w * x + bias
    error = ((y - predict)**2).mean()
    cnt += 1
    if epoch%10 == 0:
        predict = w*x + bias
        plot_prediction(predict,y)
        print("cnt=", cnt ,"epoch= ", epoch,"w= ", w, "bias= ", bias, "error= ", error)
        
    if error < 0.000001:
        break
        
```

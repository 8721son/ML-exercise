```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(100)
y = 0.3*x + 0.8
```

```python
def plot_prediction(pred,y):
    plt.plot(x,y)
    plt.plot(x,pred)
    plt.show()
```

```python
w = np.random.uniform(-1.0,1.0)
bias = np.random.uniform(-1.0,1.0)

num_epoch=100
learning_rate=0.7
cnt = 0
for epoch in range(num_epoch):
    batch_size = 10
    batch_count = 100/10
    start = 0
    end = 10
    for iteration in range(int(batch_count)):
    
        x_batch = x[start: end]
        y_batch = 0.3*x_batch + 0.8
        
        start += 10
        end += 10

        predict_batch = w*x_batch + bias
      
        w = w - 2*learning_rate*((predict_batch - y_batch)*x_batch).mean()
        bias = bias - 2*learning_rate * (predict_batch - y_batch).mean()

        predict = w*x+ bias
        error = ((y - predict)**2).mean()
        cnt += 1
        if iteration==0 :    
            plot_prediction(predict,y)
            print("cnt= ",cnt,"iteration= ", iteration,"w= ", w, "bias= ", bias, "error= ", error)
        
        # if error < 0.000001:
        #     break
```
